## 二维空间线性变换

二维线性变换必须符合以下两条规则:
1. $T(\vec x+\vec y)=T(\vec x)+T(\vec y)$: "$+$"可以拆分开
2. $T(k\vec x)=kT(\vec x)$: 可以提取系数$k$
